```python
from matplotlib.colors import to_hex,rgb2hex
def get_images(eye_dataset):
  image_array =[]
  ordered_colors_array =[]
  hex_colors_array= []
  rgb_colors_array=[]
  class_name=[]
  flattened_array=[]

  IMG_HEIGHT = 32
  IMG_WIDTH =32
  colors =[]
  for folder in os.listdir(eye_dataset):
    for file in os.listdir(os.path.join(eye_dataset,folder)):
      
      try:
        image_path = os.path.join(eye_dataset,folder,file)
        image= cv2.imread(image_path,cv2.IMREAD_COLOR)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        #hex_color = colors.to_hex(image.shape[2])
        image=np.array(image)
        #print(image.shape[2])

        try:

          ht =image.shape[0]
          wd =image.shape[1]
          cc =image.shape[2] 

          top_pad = np.floor((ht - image.shape[0]) / 2).astype(np.uint16)
          bottom_pad = np.ceil((ht - image.shape[0]) / 2).astype(np.uint16)
          right_pad = np.ceil((wd - image.shape[1]) / 2).astype(np.uint16)
          left_pad = np.floor((wd - image.shape[1]) / 2).astype(np.uint16)
          image = np.copy(np.pad(image, ((top_pad, bottom_pad), (left_pad, right_pad), (0, 0)), mode='constant', constant_values=0))

          ht =0
          wd =0
          cc =0

        except Exception as e:
            print(str(e))

        resized_image = cv2.resize(image, (IMG_HEIGHT,IMG_WIDTH), interpolation = cv2.INTER_AREA)
        reshaped_image = image.reshape(image.shape[0]*image.shape[1], 3)
        

        kmeans = KMeans(n_clusters = 2)
        labels = kmeans.fit_predict(reshaped_image)

        counts = Counter(labels)

        center_colors = kmeans.cluster_centers_
        #print(center_colors)

        # We get ordered colors by iterating through the keys
        ordered_colors = [np.round(center_colors[i]).astype(int) for i in counts.keys()]
       # round_to_whole = [round(num) for num in a_list]
        #hex_colors = [rgb2hex(ordered_colors[i]) for i in counts.keys()]
        rgb_colors = [ordered_colors[i] for i in counts.keys()]
        flattened_colors= [(ordered_colors[i]).flatten() for i in counts.keys()]


          #return rgb_colors

        
        # flat_image= reshaped_image.flatten()
          #print(image.shape)  
        
          #image = image.astype('float32')
        #print(flat_image)

      
      except Exception as e:
        print(str(e))

      
      ordered_colors_array.append(ordered_colors)
      #hex_colors_array.append(hex_colors)
      flattened_array.append(flattened_colors)
      rgb_colors_array.append(rgb_colors)
      class_name.append(folder)

  return ordered_colors_array, rgb_colors_array,class_name,flattened_array
```
